```python
import numpy as np
```

```python
import ast
```

```python
def show_array(a):
#     plt.gray()
    plt.imshow(np.squeeze(a)*255, interpolation="nearest", cmap="gray", vmin=0.0, vmax=255.0)
    plt.show()
```

```python
def maze_visits(fname, scaling=1.0, interval=None):
    with open(fname) as f:
        content = f.read().splitlines()
    visits = list(map(lambda x: ast.literal_eval(x), content))
    maze = np.zeros((28,28))
    if interval is None:
        interval = (0, len(visits))
    for visit in visits[interval[0]:interval[1]]:
        maze[visit] += 1
    gray_maze = maze / (np.max(maze) / scaling)
    gray_maze = np.clip(gray_maze,0, 1)
    show_array(gray_maze)
```

```python
fname = "Maze_4_14Cts_Logs/DQN_10_2017-02-21_02-43//logs/Player_Positions_In_Maze.txt"
for i in range(0, int(2e5), int(25e3)):
    maze_visits(fname, scaling=2, interval=(i, i+int(25e3)))
```
